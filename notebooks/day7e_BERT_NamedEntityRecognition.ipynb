# Importing Libraries

```python
from transformers import AutoTokenizer, AutoModelForTokenClassification, pipeline
```

# Initializing the Model and Tokenizer

```python
model_name = "dslim/bert-base-NER"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForTokenClassification.from_pretrained(model_name)
```

# Creating the NER Pipeline

```python
ner_pipeline = pipeline(
    "ner", model=model, tokenizer=tokenizer, aggregation_strategy="simple")
```

```python
sample_text = "Trinamool Congress leader Mahua Moitra has moved the Supreme Court against her expulsion from the Lok Sabha over the cash-for-query allegations against her. Moitra was ousted from the Parliament last week after the Ethics Committee of the Lok Sabha found her guilty of jeopardising national security by sharing her parliamentary portal's login credentials with businessman Darshan Hiranandani."
```

# Making Prediction and Error Handling

```python
results = ner_pipeline(sample_text)
if not results:
    print("No entities found in the text.")
else:
    print("\nEntities found:")
    for entity in results:
        print(
            f"- Entity: {entity['word']}, Type: {entity['entity_group']}, Position: {entity['start']}-{entity['end']}")
```
